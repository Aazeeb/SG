```python
year = [1972, 1982, 1992, 2002, 2012]
e_bangalore = [100.6, 158.61, 305.54, 394.96, 724.79]
e_mysore = [10.5, 25.21, 58.65, 119.27, 274.87]
```

```python
plt.plot(year,e_bangalore)
plt.plot(year,e_mysore)
```

```python
plt.figure(figsize=(14,6))
plt.plot(year,e_bangalore,c='r',label="Bangalore",marker="*",linestyle="dashed",linewidth=3)
plt.plot(year,e_mysore,c='k',label="Mysore",marker=".",linestyle="dashdot")
plt.legend(loc="upper left")
plt.xlabel("Years",fontsize=13)
plt.ylabel("Employed people",fontsize=14)
plt.title("Employment status",fontsize=18)
#plt.xticks(np.arange(1970,2013,2),rotation=40)
plt.xticks(year,rotation=30)
plt.grid()
plt.show()
```
